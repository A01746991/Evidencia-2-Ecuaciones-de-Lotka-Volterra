```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
ti = 0
tf = 30     
h = 0.01     
xi = 20    
yi = 20   
a = 1
b = 0.02
d = 0.01
nt = int((tf-ti)/h) 
x = np.empty(nt+1)
y = np.empty(nt+1)
x[0]= xi
y[0]= yi

for step in range(nt):
  x[step+1] = x[step]+(x[step]-(b*x[step]*y[step]))*h
  y[step+1] = y[step]+(-y[step]+(d*x[step]*y[step]))*h
```

```python
t=np.arange(nt+1) * h

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(t, x, color='green',linestyle='-', label='x')
ax.plot(t, y, color='red',linestyle='-', label='y')

ax.set_xlabel('$t$')
ax.set_ylabel('$N$')
ax.set_title('Presas/ Depredadores - Tiempo')
ax.legend()
graf1 = plt.show()
```

```python
print('Función aproximada en x:',x)
print('Función aproximada en y:',y)
print('Solución en t=',tf,':',x[nt])
print('Solución en t=',tf,':',y[nt])
```

ODEINT


```python
def model(z,t):
    dxdt = z[0]-(b*z[0]*z[1])
    dydt = -z[1]+(d*z[0]*z[1])
    dzdt = [dxdt,dydt]
    return dzdt

z0 = [20,20]

t = np.linspace(0,30,100000)

z = odeint(model,z0,t)

plt.plot(t,z[:,0],'g-',label=r'ax-bxy')
plt.plot(t,z[:,1],'r-',label=r'-cy+dxy')
plt.ylabel('Presas/Depredadores')
plt.xlabel('Tiempo')
plt.legend(loc='best')
graf2 = plt.show()
```

```python
plt.plot(z[:,0],z[:,1])
```
